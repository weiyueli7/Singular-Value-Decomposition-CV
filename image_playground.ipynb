```python
import network
import data
import image
import warnings
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
from compress import compress_image
import seaborn as sns
# sns.set_style()
```

```python
# save sample images of each class
sample_image = []
original_train_data = data.load_data(True)
for i in [0, 2, 6]:
    train_data_ind = data.np.where(original_train_data[1] == i)
    np.random.seed(100)
    train_data_ind = np.random.choice(train_data_ind[0], size = 6)
    # find the first image of its class
    selected = original_train_data[0][train_data_ind]
    sample_image.append(selected)
```

```python
for svl in [3, 8, 13]:
    fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6, figsize=(5, 1.5))
    for idx, ax in enumerate([ax1, ax2, ax3, ax4, ax5, ax6]):
        ax.imshow(compress_image(sample_image[0][idx].reshape(28, 28), svl))
        ax.set_axis_off()
#     plt.suptitle(f"Sample Class 0 with Singular Value Limit of {svl}")
    plt.tight_layout()
    plt.savefig(f"out_images/class_0_with_svd_{svl}.png", bbox_inches='tight', pad_inches=0)
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6, figsize=(5, 1.5))
for idx, ax in enumerate([ax1, ax2, ax3, ax4, ax5, ax6]):
    ax.imshow(sample_image[0][idx].reshape(28, -1))
    ax.set_axis_off()
# plt.suptitle(f"Sample Class 0 Original Images")
plt.tight_layout()
plt.savefig(f"out_images/class_0.png", bbox_inches='tight', pad_inches=0)
```
